```python
todo_list = []

def add_task():
    task = input("Enter the task: ")
    due_date = input("Enter the due date (optional): ")
    todo = {"task": task, "due_date": due_date, "completed": False}
    todo_list.append(todo)
    print("Task added successfully!\n")

def view_tasks():
    if not todo_list:
        print("No tasks in your to-do list.\n")
        return

    print("\nTo-Do List:")
    for idx, todo in enumerate(todo_list, start=1):
        status = "Done" if todo["completed"] else "Pending"
        print(f"{idx}. Task: {todo['task']} | Due Date: {todo['due_date']} | Status: {status}")
    print()

def mark_task_completed():
    view_tasks()
    task_number = int(input("Enter the task number to mark as completed: "))

    if 1 <= task_number <= len(todo_list):
        todo_list[task_number - 1]["completed"] = True
        print("Task marked as completed!\n")
    else:
        print("Invalid task number.\n")

def update_task():
    view_tasks()
    task_number = int(input("Enter the task number to update: "))

    if 1 <= task_number <= len(todo_list):
        task = input("Enter the new task description (leave blank to keep current): ").strip()
        due_date = input("Enter the new due date (leave blank to keep current): ").strip()


        if task:
            todo_list[task_number - 1]["task"] = task
        if due_date:
            todo_list[task_number - 1]["due_date"] = due_date

        print("Task updated successfully!\n")
    else:
        print("Invalid task number.\n")

def delete_task():
    view_tasks()
    task_number = int(input("Enter the task number to delete: "))

    if 1 <= task_number <= len(todo_list):
        todo_list.pop(task_number - 1)
        print("Task deleted successfully!\n")
    else:
        print("Invalid task number.\n")

def display_menu():
    print("To-Do List Application")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Mark Task as Completed")
    print("4. Update Task")
    print("5. Delete Task")
    print("6. Exit")

while True:
    display_menu()
    choice = input("Choose an option (1-6): ").strip()

    if choice == '1':
        add_task()
    elif choice == '2':
        view_tasks()
    elif choice == '3':
        mark_task_completed()
    elif choice == '4':
        update_task()
    elif choice == '5':
        delete_task()
    elif choice == '6':
        print("Exiting the program. Goodbye!")
        break
    else:
        print("Invalid choice. Please try again.\n")
```
